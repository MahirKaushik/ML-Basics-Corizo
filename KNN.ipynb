```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('diabetes.csv')
dataset.head()
```

```python
dataset.info()
```

```python
no_zero=["Glucose",	"BloodPressure","SkinThickness","Insulin","BMI"]
for i in no_zero:
    dataset[i]=dataset[i].replace(0,np.nan)
    mean=int(dataset[i].mean(skipna=True))
    dataset[i]=dataset[i].replace(np.nan,mean)

    
```

```python
dataset.head()
```

```python
#importing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#identifying datasets from training and testing

X=dataset.iloc[:,0:8]
Y=dataset.iloc[:,8]
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.2)
```

```python
# Feature Scalinf

sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
print(x_train)
```

```python
classifier=KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
